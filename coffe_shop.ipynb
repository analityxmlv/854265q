## Los datos
Se ha reunido información de tres diferentes fuentes ([locations](https://github.com/chrismeller/), [neighborhoods](http://data.denvergov.org), [demographics](https://www.census.gov/)):

#### Ubicaciones de Starbucks en Denver, Colorado
- "StoreNumber" - Número de tienda asignada por Starbucks
- "Name" - Nombre de la tienda
- "PhoneNumber" - Número telefónico de la tienda
- "Street 1, 2, and 3" - Dirección de la tienda
- "PostalCode" - Código Postal de la tienda
- "Longitude, Latitude" - Coordenadas de la tienda

#### Neighborhoods' geographical information Información geográfica de los vecindarios 
- "NBHD_ID" - Neighborhood ID (Coincide con la información de censos oficiales)
- "NBHD_NAME" - Name of the statistical neighborhood Nombre 
- "Geometry" - Forma Poligonal que define al vecindario

#### Información Demográfica
- "NBHD_ID" - Neighborhood ID (Coincide con la información Geográfica) 
- "NBHD_NAME' - Nombre del Vecindario
- "POPULATION_2010' - Población en 2010
- "AGE_ " - Número de personas por cada rango de edad (< 18, 18-34, 35-65, and > 65)
- "NUM_HOUSEHOLDS" - Número de Hogares en el vecindario
- "FAMILIES" - Número de familias en el vecindario
- "NUM_HHLD_100K+" - Número de hogares con un ingreso mayor de 100 mil USD por año
